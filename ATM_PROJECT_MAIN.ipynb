```python
import mysql.connector
conn=mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="dummy_atm"
)

cursor=conn.cursor()


create_table_query_0="""
CREATE TABLE IF NOT EXISTS atm_db(
    ACCOUNT_HOLDER_NAME VARCHAR(30),
    ACCOUNT_NO INT PRIMARY KEY ,
    PASSWORD INT
)"""

cursor.execute(create_table_query_0)


create_table_query_1="""
CREATE TABLE IF NOT EXISTS transaction_db(
    ACCOUNT_NO INT,
    TRANSACTION_ID INT PRIMARY KEY AUTO_INCREMENT,
    WITHDRAW_AMOUNT INT,
    DEPOSIT_AMOUNT INT,
    CURRENT_BALANCE INT,
    FOREIGN KEY (ACCOUNT_NO) REFERENCES atm_db(ACCOUNT_NO)
    
)"""

cursor.execute(create_table_query_1)


def insert_query_for_atm():
    ACCOUNT_HOLDER_NAME=str(input("ENTER FULL NAME: "))
    ACCOUNT_NO=int(input("ENTER ACCOUNT NUMBER: "))
    PASSWORD=int(input("ENTER PASSWORD: "))
    insert_query="""INSERT INTO atm_db(ACCOUNT_HOLDER_NAME, ACCOUNT_NO, PASSWORD) VALUES
    (%s,%s,%s)"""
    cursor.execute(insert_query,(ACCOUNT_HOLDER_NAME,ACCOUNT_NO,PASSWORD))
    conn.commit()
    print("DATA INSERTED SUCCESFULLY")

def current_balance():
    select_query_for_balance="""SELECT CURRENT_BALANCE FROM transaction_db WHERE ACCOUNT_NO=%s"""
    cursor.execute(select_query_for_balance,(USER,))
    balance=cursor.fetchone()
    
    if balance:
        print(f"Current Balance: {balance[0]}")
    else:
        print("Account number not found or no balance information available.")

def deposit_amount():
    deposit=int(input("ENTER AMOUNT TO DEPOSIT: "))

    deposit_query="""UPDATE transaction_db SET DEPOSIT_AMOUNT=%s WHERE ACCOUNT_NO=%s ;"""
    cursor.execute(deposit_query,(deposit,USER))
    DEPOSIT_AMOUNT=cursor.fetchone()
    conn.commit()
    print(f"{deposit} DEPOSITED SUCCESFULLY")
    
    CURRENT_BALANCE=""" SELECT CURRENT_BALANCE FROM transaction_db  WHERE ACCOUNT_NO=%s """
    cursor.execute(CURRENT_BALANCE,(USER,))
    show_balance=cursor.fetchone()
    dummy_balance=show_balance[0]
    dummy_balance=dummy_balance+deposit
    
    update_balance_query="""UPDATE transaction_db SET CURRENT_BALANCE=%s WHERE ACCOUNT_NO=%s ;"""
    cursor.execute(update_balance_query,(dummy_balance,USER))
    conn.commit()
    print(f"{dummy_balance} IS YOUR CURRENT BALANCE")

def withdraw_amount():
    withdraw=int(input("ENTER AMOUNT TO WITHDRAW: "))

    withdraw_query="""UPDATE transaction_db SET WITHDRAW_AMOUNT=%s WHERE ACCOUNT_NO=%s ;"""
    cursor.execute(withdraw_query,(withdraw,USER))
    WITHDRAW_AMOUNT=cursor.fetchone()
    conn.commit()
    
    
    CURRENT_BALANCE=""" SELECT CURRENT_BALANCE FROM transaction_db  WHERE ACCOUNT_NO=%s """
    cursor.execute(CURRENT_BALANCE,(USER,))
    show_balance=cursor.fetchone()
    dummy_balance=show_balance[0]
    if dummy_balance>withdraw:
        dummy_balance=dummy_balance-withdraw
        print(f"{withdraw} WITHDRAWN SUCCESFULLY")
        
    else:
        print("INSUFFICIENT BALANCE")
    update_withdraw_query="""UPDATE transaction_db SET CURRENT_BALANCE=%s WHERE ACCOUNT_NO=%s ;"""
    cursor.execute(update_withdraw_query,(dummy_balance,USER))
    conn.commit()
    print(f"{dummy_balance} IS YOUR CURRENT BALANCE")


    

USER=int(input("ENTER ACCOUNT NUMBER: "))
PASSW=int(input("ENTER PASSWORD: "))
select_query=""" SELECT * FROM atm_db WHERE ACCOUNT_NO=%s and PASSWORD=%s """
cursor.execute(select_query,(USER,PASSW))
LOGIN_TRY=cursor.fetchone()
print
if LOGIN_TRY:
    print("LOGIN SUCCESFULL")
    choose=input("ENTER YOUR CHOICE FROM BELOW OPTIONS:\n\t1.CHECK CURRENT BALANCE\n\t2.DEPOSIT AMOUNT\n\t3.WITHDRAW AMOUNT")
    choose=choose.upper()
    if choose=="CHECK":
        current_balance()
    elif choose=="DEPOSIT":
        deposit_amount()
    elif choose=="WITHDRAW":
        withdraw_amount()
    else:
        print("INPUT ERROR! TRY AGAIN")
        
        
    
else:
    print("INAPPROPRIATE INPUT! TRY AGAIN.")


    



```
